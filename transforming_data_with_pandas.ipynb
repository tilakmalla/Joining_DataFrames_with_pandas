## Importing important libraries into the environment

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import pandas as pd
import numpy as np
```

## Importing data into DataFrame object 

```python
main_data = pd.read_csv("ted_main.csv") 
```

## Selecting top two rows

```python

main_data.head(2)
```

## A few columns are selected

```python

mutated_data = main_data[['title', 'main_speaker','duration', 'views', 'comments']]

mutated_data.head(2)
```

## Sort values by column 'views' descending order`

```python
mutated_data.sort_values('views', ascending =False)
```

## Number of comments received per 10 thousand views 

```python
mutated_data.loc[:,'comments_per_10k_views'] \
            =10000 * mutated_data.loc[:,'comments']/mutated_data.loc[:,'views']
```

## Top Provoking Titles 
* Topic that received more comments per view is assumed more provoking 

```python
mutated_data.sort_values('comments_per_10k_views', ascending =False).head()
```

## All the titles receiving above 10 comments per 10k views

```python
mutated_data[mutated_data['comments_per_10k_views'] > 10]\
                .sort_values('comments_per_10k_views', ascending =False)
```
